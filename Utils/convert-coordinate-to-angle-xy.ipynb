```python
#ìž…ë ¥ë¶€
points_x = []
points_y = []

user_input = "" # ì‚¬ìš©ìž ìž…ë ¥
while True:
    user_input = input("xì¢Œí‘œ, yì¢Œí‘œ, xë„, yë„ë¥¼ ìž…ë ¥í•˜ì„¸ìš” (ì˜ˆ: 10 5 90 100), 0 to quit >>> ")
    
    if user_input == "0":
        break
    
    temp = list(map(int, user_input.split()))
    
    temp_x = temp[:3]
    temp_y = [temp[0], temp[1], temp[3]]
    
    points_x.append(temp_x)
    points_y.append(temp_y)

# xì¢Œí‘œ, yì¢Œí‘œ, xë„
```

```python
# ëª¨ë“  ì¡°í•©
from itertools import combinations

per_list_x = list(combinations(points_x, 3))
per_list_y = list(combinations(points_y, 3))
per_list_y
```

```python
import numpy as np

rstk_x = []
rstk_y = []

for data in per_list_x:
    # v1 = (ð‘Ž1,ð‘1,ð‘1)âˆ’(ð‘Ž2,ð‘2,ð‘2) = (ð‘Ž1âˆ’ð‘Ž2,ð‘1âˆ’ð‘2,ð‘1âˆ’ð‘2)
    v1 = (data[0][0]-data[1][0], data[0][1]-data[1][1], data[0][2]-data[1][2])
    # v2 = (ð‘Ž1,ð‘1,ð‘1)âˆ’(ð‘Ž3,ð‘3,ð‘3)=(ð‘Ž1âˆ’ð‘Ž3,ð‘1âˆ’ð‘3,ð‘1âˆ’ð‘3)
    v2 = (data[0][0]-data[2][0], data[0][1]-data[2][1], data[0][2]-data[2][2])
    
    # ð§=ð¯1Ã—ð¯2=(ð‘Ÿ,ð‘ ,ð‘¡) cross product
    r, s, t = np.cross(v1, v2)
    
    # ð‘˜=ð‘Ÿð‘Ž1+ð‘ ð‘1+ð‘¡ð‘1
    k = r*data[0][0] + s*data[0][1] + t*data[0][2]
    
    rstk_x.append((r, s, t, k))
    
    
for data in per_list_y:
    # v1 = (ð‘Ž1,ð‘1,ð‘1)âˆ’(ð‘Ž2,ð‘2,ð‘2) = (ð‘Ž1âˆ’ð‘Ž2,ð‘1âˆ’ð‘2,ð‘1âˆ’ð‘2)
    v1 = (data[0][0]-data[1][0], data[0][1]-data[1][1], data[0][2]-data[1][2])
    # v2 = (ð‘Ž1,ð‘1,ð‘1)âˆ’(ð‘Ž3,ð‘3,ð‘3)=(ð‘Ž1âˆ’ð‘Ž3,ð‘1âˆ’ð‘3,ð‘1âˆ’ð‘3)
    v2 = (data[0][0]-data[2][0], data[0][1]-data[2][1], data[0][2]-data[2][2])
    
    # ð§=ð¯1Ã—ð¯2=(ð‘Ÿ,ð‘ ,ð‘¡) cross product
    r, s, t = np.cross(v1, v2)
    
    # ð‘˜=ð‘Ÿð‘Ž1+ð‘ ð‘1+ð‘¡ð‘1
    k = r*data[0][0] + s*data[0][1] + t*data[0][2]
    
    rstk_y.append((r, s, t, k))
    
```

```python
# x
x_sum_r = 0
x_sum_s = 0
x_sum_t = 0
x_sum_k = 0

for data in rstk_x:
    x_sum_r += data[0]
    x_sum_s += data[1]
    print(x_sum_r)
    x_sum_t += data[2]
    x_sum_k += data[3]

x_avg_r = x_sum_r / len(rstk_x)
x_avg_s = x_sum_s / len(rstk_x)
x_avg_t = x_sum_t / len(rstk_x)
x_avg_k = x_sum_k / len(rstk_x)

# y
y_sum_r = 0
y_sum_s = 0
y_sum_t = 0
y_sum_k = 0

for data in rstk_y:
    y_sum_r += data[0]
    y_sum_s += data[1]
    y_sum_t += data[2]
    y_sum_k += data[3]

y_avg_r = y_sum_r / len(rstk_y)
y_avg_s = y_sum_s / len(rstk_y)
y_avg_t = y_sum_t / len(rstk_y)
y_avg_k = y_sum_k / len(rstk_y)
```
